```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', parse_dates=['timestamp'])
```

```python
print(df.head())
```

```python
print(df.select_dtypes(include='number').describe())
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['timestamp'], df['posture_angle_deg'], label='Posture Angle')
plt.plot(df['timestamp'], df['target_posture_angle_deg'], label='Target Angle', linestyle='--')
plt.xticks(rotation=45)
plt.ylabel('Degrees')
plt.title('Posture vs Target Across the Day')
plt.legend()
plt.tight_layout()
```
